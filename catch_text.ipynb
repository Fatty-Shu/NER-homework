```python
import requests
from bs4 import BeautifulSoup

def scrape_agriculture_news():
    """爬取农业新闻网站数据"""
    urls = [
        "https://www.agri.cn/",        # 中国农业网
        "https://www.farmer.com.cn/",  # 农民日报
        "https://www.zgny.com.cn/"     # 中国农业新闻网
    ]
    
    for url in urls:
        try:
            response = requests.get(url, timeout=5)
            soup = BeautifulSoup(response.text, 'html.parser')
            # 提取文本内容
            texts = [p.get_text() for p in soup.find_all('p')]
            # 保存到文件
            with open('agriculture_texts.txt', 'a', encoding="gbk") as f:
                f.write('\n'.join(texts))
        except:
            continue
```

```python
scrape_agriculture_news()
```
